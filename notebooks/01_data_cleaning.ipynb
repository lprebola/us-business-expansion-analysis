```python
#Import Languages
import pandas as pd
import numpy as np
```

```python
#Read excel files
unemp_df = pd.read_excel("../data/raw/laucnty24.xlsx", skiprows=4)
pop_df = pd.read_excel("../data/raw/NST-EST2024-POP.xlsx", skiprows=3)
income_df = pd.read_excel("../data/raw/census_income.xlsx")
```

```python
unemp_df.columns = unemp_df.columns.astype(str)
pop_df.columns = pop_df.columns.astype(str)
income_df.columns = income_df.columns.astype(str)
```

```python
#Clean Column names
unemp_df.columns = [col.strip().replace(' ', '_').lower() for col in unemp_df.columns]
pop_df.columns = [col.strip().replace(' ', '_').lower() for col in pop_df.columns]
income_df.columns = [col.strip().replace(' ', '_').lower() for col in income_df.columns]
```

```python
#Prints the (Rows, Columns) # in each file
print(unemp_df.shape)
print(pop_df.shape)
print(income_df.shape)
```

```python
#Prints the names of each column in each list
print(unemp_df.columns.tolist())
print(pop_df.columns.tolist())
print(income_df.columns.tolist())
```

```python
#Rename unemployment Columns
unemp_df.columns = ["code", "state_fips", "county_fips", "county_name",
                    "year", "labor_force", "employed", "unemployed", "unemployment_rate"]
unemp_df.head()
```

```python
pop_df.columns = ["state", "base_2020", "2020", "2021", "2022", "2023", "2024"]
pop_df.head()
```

```python
income_df = income_df.drop(columns=["state.1"])
income_df.head()
```

```python
#Gives me the missing items in each column
print(unemp_df.isnull().sum())
print(pop_df.isnull().sum())
print(income_df.isnull().sum())
```

```python
#Remove all the unwanted regions and Puerto Rice, and fix names so theres no . at start, also drop all the nulls
pop_df["state"] = pop_df["state"].str.strip().str.replace(".", "", regex=False)
drop_rows = ["United States", "Northeast", "Midwest", "South", "West", "Puerto Rico", None]
pop_df = pop_df[~pop_df["state"].isin(drop_rows)]
pop_df = pop_df[["state", "2024"]].rename(columns={"2024": "population"})
pop_df["population"] = pd.to_numeric(pop_df["population"], errors="coerce")
pop_df = pop_df.dropna()
pop_df.head()
```

```python
pop_df.shape
```

```python
#filter and keep every row but where the State is Puerto Rico
income_df = income_df[income_df["state"] != "Puerto Rico"]
print(income_df.shape)
income_df.head()
```

```python
#Extract state abbreviation from county Name
unemp_df["state_abbr"] = unemp_df["county_name"].str.extract(r",\s*([A-Z]{2})$")
#Group by state and average unemplyment across all counties
unemp_state = unemp_df.groupby("state_abbr")["unemployment_rate"].mean().reset_index()
unemp_state["unemployment_rate"] = unemp_state["unemployment_rate"].round(2)

print(unemp_state.shape)
unemp_state.head()
```

```python
#Map state abbreviations to full names
state_abbr_map = {
    "AL": "Alabama", "AK": "Alaska", "AZ": "Arizona", "AR": "Arkansas",
    "CA": "California", "CO": "Colorado", "CT": "Connecticut", "DE": "Delaware",
    "DC": "District of Columbia", "FL": "Florida", "GA": "Georgia", "HI": "Hawaii",
    "ID": "Idaho", "IL": "Illinois", "IN": "Indiana", "IA": "Iowa", "KS": "Kansas",
    "KY": "Kentucky", "LA": "Louisiana", "ME": "Maine", "MD": "Maryland",
    "MA": "Massachusetts", "MI": "Michigan", "MN": "Minnesota", "MS": "Mississippi",
    "MO": "Missouri", "MT": "Montana", "NE": "Nebraska", "NV": "Nevada",
    "NH": "New Hampshire", "NJ": "New Jersey", "NM": "New Mexico", "NY": "New York",
    "NC": "North Carolina", "ND": "North Dakota", "OH": "Ohio", "OK": "Oklahoma",
    "OR": "Oregon", "PA": "Pennsylvania", "RI": "Rhode Island", "SC": "South Carolina",
    "SD": "South Dakota", "TN": "Tennessee", "TX": "Texas", "UT": "Utah",
    "VT": "Vermont", "VA": "Virginia", "WA": "Washington", "WV": "West Virginia",
    "WI": "Wisconsin", "WY": "Wyoming", "DC": "District of Columbia"
    }

unemp_state["state"] = unemp_state["state_abbr"].map(state_abbr_map)
unemp_state = unemp_state.dropna(subset=["state"])

#Merge all three
merged_df = income_df.merge(pop_df, on="state")
merged_df = merged_df.merge(unemp_state[["state", "unemployment_rate"]], on="state")

print(merged_df.shape)
merged_df.head()
```

```python
merged_df.to_csv("../data/cleaned/merged_state_data.csv", index=False)
print("Saved to data/cleaned/merged_state_data.csv")
```
